```python
from konlpy.tag import Okt
import pandas as pd

common_path = "/content/drive/MyDrive/기학_팀플젝/박채연/샘플링/"
save_path = "/content/drive/MyDrive/기학_팀플젝/박채연/tokenized/"
```

```python
okt = Okt()
```

```python
exclude_tags = {'Josa', 'Eomi', 'PreEomi', 'Punctuation', 'Exclamation',
                'Foreign', 'Unknown', 'KoreanParticle', 'Hashtag',
                'ScreenName', 'Alpha'}
```

```python
def tokenize(filename):
    df = pd.read_csv(common_path + filename)

    for i in range(len(df)):
        # newsTitle 처리
        title = df["newsTitle"][i]
        title_pos = okt.pos(title, norm=True, stem=True)
        filtered_title_tokens = [word for word, tag in title_pos if tag not in exclude_tags]
        df.at[i, "newsTitle"] = " ".join(filtered_title_tokens)

        # newsContext 처리
        context = df["newsContent"][i]
        context_pos = okt.pos(context, norm=True, stem=True)
        filtered_context_tokens = [word for word, tag in context_pos if tag not in exclude_tags]
        df.at[i, "newsContent"] = " ".join(filtered_context_tokens)

    df.to_csv(save_path+filename, index=False)
```

```python
tokenize("sampled_extracted_data_Clickbait_PO.csv")
```
