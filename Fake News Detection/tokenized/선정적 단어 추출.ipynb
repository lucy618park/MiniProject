```python
from collections import Counter
import pandas as pd
from konlpy.tag import Okt
```

```python
okt = Okt()
```

```python
def process_and_count(filenames, common_path):
    all_tokens = []


    for filename in filenames:
        df = pd.read_csv(common_path + filename)
        filtered_df = df[df["processPattern"] == 1]

        for tokenized_title in filtered_df["newsTitle"]:
            tokens = tokenized_title.split() if isinstance(tokenized_title, str) else []
            all_tokens.extend(tokens)

    word_counts = Counter(all_tokens)

    result_df = pd.DataFrame(word_counts.items(), columns=["Word", "Frequency"])
    result_df = result_df.sort_values(by="Frequency", ascending=False).reset_index(drop=True)

    result_df.to_csv(common_path + "선정적 단어 추출.csv", index=False)
```

```python
common_path = "/content/drive/MyDrive/기학_팀플젝/박채연/tokenized/"
filenames = ["sampled_extracted_data_Clickbait_GB_D.csv", "sampled_extracted_data_Clickbait_IS_D.csv",
             "sampled_extracted_data_Clickbait_PO.csv", "sampled_extracted_data_Clickbait_SO.csv"]
```

```python
process_and_count(filenames, common_path)
```
