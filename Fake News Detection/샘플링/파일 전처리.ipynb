```python
import pandas as pd

common_path = "/content/drive/MyDrive/기학_팀플젝/박채연/"
```

```python
#계층화 샘플링 과정(상, 중, 하를 기준으로 각 20%씩 추출)
def sample_data(filename):
  df = pd.read_csv(common_path+'json 파싱/'+filename)

  sampled_df = df.groupby('processLevel', group_keys=False).apply(lambda x: x.sample(frac=0.2, random_state=42))
  #processPattern 값이 13이나 14면 1, 나머지는 0으로 변경
  sampled_df['processPattern'] = sampled_df['processPattern'].apply(lambda x: 1 if x in [13, 14] else 0)

  sampled_df.to_csv(common_path+'샘플링/sampled_'+filename, index=False)
```

```python
sample_data("extracted_data_Clickbait_PO.csv")
sample_data("extracted_data_Clickbait_SO.csv")
sample_data("extracted_data_NonClickbait_EC.csv")
sample_data("extracted_data_NonClickbait_ET.csv")
sample_data("extracted_data_NonClickbait_GB.csv")
sample_data("extracted_data_Clickbait_GB_D.csv")
sample_data("extracted_data_Clickbait_IS_D.csv")
```
