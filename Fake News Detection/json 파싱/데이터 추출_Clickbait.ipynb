```python
import zipfile
import os
import json
import pandas as pd
```

```python
def extract_json_from_zip(zip_path, target_keys=None):
    extracted_data = []

    # zip 파일 열기
    with zipfile.ZipFile(zip_path, 'r') as z:
        # zip 파일 안의 파일 목록 중 .json 파일만 추출
        json_files = [f for f in z.namelist() if f.endswith('.json')]

        for json_file in json_files:
            # 각 JSON 파일 열기
            with z.open(json_file) as f:
                data = json.load(f)  # JSON 파일 읽기
                source_info = data.get("sourceDataInfo", {})
                source_info2 = data.get("labeledDataInfo", {})

                # 필요한 키 값만 추출하여 딕셔너리로 구성
                extracted_row = {
                    "newsId": source_info.get("newsID"),
                    "newsTitle": source_info2.get("newTitle"),
                    "newsContent": source_info.get("newsContent"),
                    "useType": source_info.get("useType"),
                    "processLevel": source_info.get("processLevel"),
                    "processPattern": source_info.get("processPattern"),
                }

                extracted_data.append(extracted_row)

        df = pd.DataFrame(extracted_data)
        return df
```

```python
zip_path = '/content/drive/MyDrive/기학_팀플젝/데이터/가공 전 데이터/TL_Part1_Clickbait_Auto_PO.zip'
target_keys = ['newsId', 'newsTitle', 'newsContent', 'useType', 'processLevel', 'processPattern']
df = extract_json_from_zip(zip_path, target_keys)

# DataFrame을 CSV 파일로 저장
df.to_csv('/content/drive/MyDrive/기학_팀플젝/박채연/extracted_data_Clickbait_PO.csv', index=False)
print("CSV 파일로 저장 완료!")
```
